```python
from scipy.interpolate import interp1d
import numpy as np

# 주어진 데이터 포인트
x = np.array([0, 1, 2, 3, 4, 5, 10, 11])
y = np.array([0, 0.8, 0.9, 0.1, -0.8, -1, 0, -0.2])

# interp1d 객체 생성
f = interp1d(x, y, kind='cubic')

# 새로운 x 값들에서 보간된 y 값들 생성
xnew = np.linspace(0, 11, num=100, endpoint=True)
ynew = f(xnew)

# 사용 예: xnew와 ynew를 이용하여 그래프 그리기 등
```

```python
from matplotlib import pyplot as plt
```

```python
plt.scatter(xnew, ynew)
plt.show()
```
